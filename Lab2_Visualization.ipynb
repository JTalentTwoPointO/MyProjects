###19.Create the same chart from Exercise 8 using your custom style

```python
plt.style.use('styler.mplstyle')

data = pd.read_csv(link)
pie = data.iloc[:,1:7]
s = pie.sum()
def auto_pct_func(pct, vals):
    abs = int(pct / 100.*np.sum(vals))
    return "{:.1f}%\n".format(pct, abs)
l = [str(x) for x in pie]
plt.pie(s,labels=l, shadow=True,autopct = lambda PCT: auto_pct_func(PCT, x))
plt.legend(bbox_to_anchor=(1,0.45),loc='upper center')
plt.title('Sales data')
plt.show()
```

####21.Create two box plots to compare the earnings of recent college graduates in majors thatare primarily composed of women to the earnings of recent college graduates in majors that are primarily composed of men.
*   a) Does there appear to be a significant difference? (Hint: Use notches.)
*   b) Remember to label the categories!



```python
data = pd.read_csv('https://pastebin.com/raw/cKkZLMKy', index_col='Rank') #Collage majors csv uploaded to pastebin
data
m = data['Major'][data['Men'].idxmax()] 
w = data['Major'][data['Women'].idxmax()]
data = data.loc[data['Major'].isin([m,w])]

df = data[['Major',"Men","Women"]]
tidy = df.melt(id_vars=['Major'],value_vars=['Men','Women'])
sb.boxplot(x="Major",y='value',hue='variable',data=tidy)
plt.show()
```

###23.Use small multiples to compare the trends in gender ratio of 17 U.S. college majors between 1970 and 2011. Make sure all of the subplots have the same axes and ranges.
a. Only label the y-axes of the subplots on the left  
b. only label the x-axes of the subplots on the bottom.

```python

from bokeh.layouts import *
output_notebook()

data = pd.read_csv('https://pastebin.com/raw/WXp2U1dr') #woman data
data
year = data['Year']

plots = []
n = int(data.shape[1]**0.5)
for col in data.iloc[:,1:]:
  s1 = figure(width=250, plot_height=250, title=col)
  s1.circle(year,data[col], size=5, color="navy", alpha=0.5)
  s1.xaxis.visible = data.columns.get_loc(col) + n >= data.shape[1]
  s1.yaxis.visible = (data.columns.get_loc(col) % n) == 1
  plots.append( s1)

# show the results
show(gridplot(plots, ncols=n))
```
